### ALPG

```python
import pandas as pd

df1 = pd.read_excel('../data/alpg-min.xlsx')
#print(df.head())
df1.tail()
```

```python
## Akumulasi menit ke jam

perjam1 = []
jdh = 60
for xx in range (0, 525600, 60):
    jum1 = df1['minutes'][xx:xx+jdh].sum()
    perjam1.append(jum1)
    #print(zz)
```

```python
#kumpulkan semua yang sama jam, akumulasi, rerata Winter
dfr1w1 = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (1,1416):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24
    dfr1w1.append(jum/59)       # rata-rata
    jum = 0
print(hari)

dfr1w2 = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (8016,8760):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1w2.append(jum/31)     # rata-rata
    jum = 0
print(hari)

dfr1w = [sum(x)/2 for x in zip(dfr1w2, dfr1w1)]

#import matplotlib.pyplot as plt 
#plt.plot(dfr1w)
```

```python
dfr1w
```

```python
#kumpulkan semua yang sama jam, akumulasi, rerata Spring

dfr1sp = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (1416,3624):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1sp.append(jum/92)     # rata-rata
    jum = 0
print(hari)
#import matplotlib.pyplot as plt 
#plt.plot(dfr1sp)
```

```python
#kumpulkan semua yang sama jam, akumulasi, rerata Summer

dfr1su = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (3624,5832):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1su.append(jum/92)     # rata-rata
    jum = 0
print(hari)
#import matplotlib.pyplot as plt 
#plt.plot(dfr1su)
```

```python
#kumpulkan semua yang sama jam, akumulasi, rerata Autum

dfr1au = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (5832,8016):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1au.append(jum/91)     # rata-rata
    jum = 0
print(hari)
#import matplotlib.pyplot as plt 
#plt.plot(dfr1au)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(20,10), dpi=300)
fig.subplots_adjust(bottom=0.025, left=0.025, top = 0.975, right=0.975)

plt.subplot(2,2,1)
plt.gca().grid(True)
plt.plot(dfr1w, color="green", linewidth=2, label="ALPG-Winter")
plt.legend(loc='upper left', frameon=True)
plt.ylabel('Share(%)', fontsize=18) 
plt.xticks(range(24))
#plt.xlabel('Time')
plt.tick_params(which='both',labelbottom=True)

plt.subplot(2,2,2)
plt.gca().grid(True)
plt.plot(dfr1sp, color="blue", linewidth=2, label="ALPG-Spring")
plt.legend(loc='upper left', frameon=True)
plt.ylabel('Share(%)', fontsize=18) 
plt.xticks(range(24))
#plt.xlabel('Time')
plt.tick_params(which='both',labelbottom=True)

plt.subplot(2,2,3)
plt.gca().grid(True)
plt.plot(dfr1su, color="red", linewidth=2, label="ALPG-Summer")
plt.legend(loc='upper left', frameon=True)
plt.ylabel('Share(%)', fontsize=18)    
plt.xlabel('Time', fontsize=18)
plt.xticks(range(24))
plt.tick_params(which='both',labelbottom=True)

plt.subplot(2,2,4)
plt.gca().grid(True)
plt.plot(dfr1au, color="green", linewidth=2, label="ALPG-Autumn")
plt.legend(loc='upper left', frameon=True)
plt.ylabel('Share(%)', fontsize=18)    
plt.xlabel('Time', fontsize=18)
plt.xticks(range(24))
plt.tick_params(which='both',labelbottom=True)

plt.savefig('ALPG Season.png', dpi=500, facecolor='w', edgecolor='w',
        orientation='portrait', papertype='legal', format=None,
        transparent=False, bbox_inches='tight', pad_inches=0.1,
        frameon=True)

plt.show()
```
