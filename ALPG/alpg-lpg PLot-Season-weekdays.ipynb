```python
## ALPG

import pandas as pd

df1 = pd.read_excel('../data/alpg-min.xlsx')

## Akumulasi menit ke jam
perjam1 = []
jdh = 60
for xx in range (0, 525600, 60):
    jum1 = df1['minutes'][xx:xx+jdh].sum()
    perjam1.append(jum1)

#kumpulkan semua yang sama jam, akumulasi, rerata Winter
dfr1w1 = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (0,1416):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24
    dfr1w1.append(jum/59)       # rata-rata
    jum = 0

dfr1w2 = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (8016,8760):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1w2.append(jum/31)     # rata-rata
    jum = 0

dfr1w = [sum(x)/2 for x in zip(dfr1w2, dfr1w1)]

#kumpulkan semua yang sama jam, akumulasi, rerata Spring

dfr1sp = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (1416,3624):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1sp.append(jum/92)     # rata-rata
    jum = 0

#kumpulkan semua yang sama jam, akumulasi, rerata Summer

dfr1su = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (3624,5832):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1su.append(jum/92)     # rata-rata
    jum = 0

#kumpulkan semua yang sama jam, akumulasi, rerata Autum

dfr1au = []
jum = 0
ite1 = 0
for yy in range (0,24):
    for xx in range (5832,8016):
        if xx % 24 == yy:       # kumpulkan semua yang sama jam
            jum += perjam1[xx]  # Akumulasi
            ite1+=1
    hari =  ite1/24             #hari
    dfr1au.append(jum/91)     # rata-rata
    jum = 0
```
